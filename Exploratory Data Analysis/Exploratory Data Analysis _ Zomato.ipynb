```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country, on = 'Country Code', how = 'left')
final_df.head()
```

```python
final_df.dtypes 
```

```python
final_df.columns
```

```python
country_name = final_df.Country.value_counts().index # value_counts() it will group by the country and returns count(*)
```

```python
country_value = final_df.Country.value_counts().values
```

```python
plt.pie(country_value[0:3], labels = country_name[0:3], autopct = '%1.2f%%')
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'Rating_count'})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating_count', data = ratings)  
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating_count', hue = 'Rating color', data = ratings, palette = ['white', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
# Plotting the categorical variable
sns.countplot(x = 'Rating color', data = ratings, palette=['blue','red', 'orange', 'yellow'])
# It plots the frequency
```

```python
# Which countries has given zero rating and for how many times?
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

```python
final_df.groupby(['Aggregate rating', 'Country']).size().reset_index()
```

```python
# Currency used by country
final_df.groupby(['Country', 'Currency']).size().reset_index()
```

```python
# Which countries have online delievery
final_df[['Has Online delivery', 'Country']].groupby(['Has Online delivery', 'Country']).size().reset_index()
```
